## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```python
# (A)
qtd_total = 3
ticket_medio = 320.52
valor_total = qtd_total * ticket_medio
print(valor_total)
```

```python
# (B)
ticket_medio = 119.21
valor_total = 834.47
qtd_total = int(valor_total / ticket_medio)
print(qtd_total)
```

```python
# (C)
qtd_total = 5
valor_total = 15378.12
ticket_medio =  valor_total / qtd_total
print(f'{ticket_medio:.2f}')
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
cancao = cancao.replace('pião', 'pavão')
cancao = cancao[0:10].upper() + cancao[10:]
moinho = cancao.find('moinho')
print(moinho)
print(cancao)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

onde_selic = noticia.find('%')
selic = noticia[onde_selic-4:onde_selic]
print(selic)

onde_anos = noticia.find('anos')
anos = noticia[onde_anos-2:onde_anos]
print(anos)
```
