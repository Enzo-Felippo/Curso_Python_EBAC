## 1\. Excel para CSV

Utilizando o pacote Python `openpyxl` visto em aula, extraia os seguintes as colunas `id`, `sexo` e `idade` para dos clientes inadimplentes (`default = 1`) e solteiros (`estado_civil = 'solteiro'`). Salves os dados extraídos no arquivo csv `credito.csv` separado por `;`. Exemplo do cabeçalho e das três primeiras linhas:

```
id;sexo;idade
767712558;59;M
713741358;46;M
772390908;59;M
```

**Dica:** O arquivo csv `credito.csv` deve ter 669 linhas, contando com o cabeçalho.

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````python
from openpyxl import load_workbook, Workbook

planilhas = load_workbook(filename='.\Consumo\credito.xlsx')
planilha = planilhas.active

def estracao_dados(indice, filtro, condicao) -> list:
    cabecalho = next(planilha.values)
    coluna = cabecalho.index(indice)
    filtrar = cabecalho.index(filtro)
    colunas = [linha[coluna] for linha in planilha.values if linha[filtrar] == condicao]
    return colunas
````

````python
import csv
def escrever_arquivo_csv(indice, dados):
    with open(file='.\Produto\inadiplentes_solteiros.csv', mode='w', encoding='utf8', newline='') as arq:
        escrever = csv.writer(arq)
        escrever.writerow([indice])
        for dado in dados:
            escrever.writerow([dado])

inadiplentes = estracao_dados(indice='id', filtro='default', condicao=1)
solteiros = estracao_dados(indice='id', filtro='estado_civil', condicao='solteiro')
filtro = set(solteiros).intersection(inadiplentes)
escrever_arquivo_csv('id', filtro)

````

````python
import json
from openpyxl import Workbook

def estracao_dados(indice) -> list:
    cabecalho = next(planilha.values)
    coluna = cabecalho.index(indice)
    colunas = [linha[coluna] for linha in planilha.values]
    return list(set(colunas[1:]))

tipo_cartao = estracao_dados(indice='tipo_cartao')
escolaridade = estracao_dados(indice='escolaridade')
credito = {'tipo_cartao':tipo_cartao, 'escolaridade':escolaridade }

credito_json = json.dumps(credito, indent=4)
print(credito_json)

with open(file='.\Produto\dados.json', mode='w') as arq:
    json.dump(obj=credito, fp=arq, indent=4)
````

## 3\. BÔNUS: Texto para CSV

No arquivo de texto `ebac.txt` você encontra o texto presente no rodapé da página de cursos da EBAC ([link](https://ebaconline.com.br/)).



**Arquivo TXT:** ebac.txt

````python
%%writefile Consumo/ebac.txt
MÍDIAS SOCIAIS
Instagram, Facebook, Youtube, LinkedIn 

CURSOS
Software, Design, Marketing, Audiovisual, Programação & Data, Games

WEBINARS
Próximos, Anteriores

SOBRE
Sobre nós, Centro de carreiras, Vagas

CONTATO
WhatsApp +55 (11) 4200-2991
Telefone +55 (11) 3030-3200

BLOG
Design, Audiovisual, Marketing
````

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
telefone;+551130303200
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````python
import re
import csv

def abrir_arquivo():
    with open(file='.\Consumo\ebac.txt', mode='r', encoding='utf8') as arq:
        return arq.read()

def filtro_numero(arquivo: list) -> list:
    return re.findall('\+\d{2}\s?\(\d{2}\)\s?\d{4,5}-\d{4}', arquivo)

def filtro_linha_numero(arquivo: list) -> list:
    return re.findall('(.+)\d', arquivo)

def limpar_numero(arquivo: list) -> list:
    return [re.sub('[\(\)\s-]', '', linha) for linha in arquivo]

def limpar_texto(arquivo: list) -> list:
    arquivo = limpar_numero(arquivo)
    return [linha.split('+')[0] for linha in arquivo]

def escrever_arquivo(indice, conteudo):
    with open(file='.\Produto\contato_ebac.csv', mode='w', encoding='utf8', newline='') as arq:
        escrever = csv.writer(arq, delimiter=';')
        for pocisao in range(len(indice)):
            escrever.writerow([indice[pocisao], conteudo[pocisao]])

arquivo = abrir_arquivo()
arquivo_texto = limpar_texto(filtro_linha_numero(arquivo))
arquivo_numero = limpar_numero(filtro_numero(arquivo))
escrever_arquivo(arquivo_texto, arquivo_numero)
````
